## Load data

```python
import sys
from os import path
sys.path.append(path.pardir)
from src.covid import CovidCountry
import pandas as pd

# Get our Country model data
country = 'Brazil'
filepath = '../data/raw/'

cty = CovidCountry(country=country,
                  filepath=filepath)
cty.load()
```

## Fit and Predict

```python
cty.fit(curve='gamma')
df = cty.predict()
cty.save(filepath='../data/processed/')
```
