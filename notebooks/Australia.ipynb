## Australia Data
## Load data

```python
import sys
from os import path
sys.path.append(path.pardir)
from src.covid import CovidCountry
import pandas as pd

# Load data
country = 'Australia'
filepath = '../data/raw/'
fix_data = {}

au = CovidCountry(country=country,
                  filepath=filepath,
                  fixes=fix_data)

au.load(today=6440)
```

## Fit Model, Make Predictions

```python
au.fit(curve='gamma')
au_df = au.predict()
au.save(filepath='../data/processed/')
```

```python
# Show small window of prediction and actual
au_df[pd.Timestamp.now().date()-pd.Timedelta('3D'):].diff()[1:6]
```
