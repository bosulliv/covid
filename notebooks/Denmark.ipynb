## Denmark Data
## Load data, Fit Model

```python
import sys
from os import path
sys.path.append(path.pardir)
from src.covid import CovidCountry
import pandas as pd

# Load data
country = 'Denmark'
filepath = '../data/raw/'
fix_data = {}

dk = CovidCountry(country=country,
                  filepath=filepath,
                  fixes=fix_data)

dk.load()
```

```python
dk.fit(curve='gamma')
dk_df = dk.predict()
dk.save(filepath='../data/processed/')
```

```python
# Show small window of prediction and actual
dk_df[pd.Timestamp.now().date()-pd.Timedelta('3D'):].diff()[1:6]
```
